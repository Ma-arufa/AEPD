```python
import pandas as pd
```

```python
AEPD_new = pd.read_excel(r'C:\Users\dell\Documents\AEPD.xlsx')
print (AEPD_new)
```

```python
#Firstly, we normalise our dataset to a common scale using the min max scaler 
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler() 
normalised_AEPD_new = pd.DataFrame(scaler.fit_transform(AEPD_new), columns=AEPD_new.columns) 
features_AEPD_new = normalised_AEPD_new
heating_target = normalised_AEPD_new['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_AEPD_new, heating_target, test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
AEPD_linear = LinearRegression()
#fit the model to the training dataset
AEPD_linear.fit(x_train, y_train)
#obtain predictions
predicted_values = AEPD_linear.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 20)
```

```python
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 30)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse, 20)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
#ridge/l2

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1.0) 
ridge_reg.fit(x_train, y_train)
```

```python
#lasso/l1

from sklearn.linear_model import Lasso 
lasso_reg = Lasso(alpha=1.0) 
lasso_reg.fit(x_train, y_train)
```

```python
#comparing the effects of regularisation
def get_weights_AEPD_new(model, feat, col_name): 
    #this function returns the weight of every feature 
    weights = pd.Series(model.coef_, feat.columns).sort_values() 
    weights_AEPD_new = pd.DataFrame(weights).reset_index()  
    weights_AEPD_new.columns = ['Features', col_name]  
    weights_AEPD_new[col_name].round(3)  
    return weights_AEPD_new
```

```python
AEPD_linear_weights = get_weights_AEPD_new(AEPD_linear, x_train, 'AEPD_linear_Weight') 
ridge_weights_AEPD_new = get_weights_AEPD_new(ridge_reg, x_train, 'Ridge_Weight') 
lasso_weights_AEPD_new = get_weights_AEPD_new(lasso_reg, x_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(AEPD_linear_weights, ridge_weights_AEPD_new, on='Features') 
final_weights = pd.merge(final_weights, lasso_weights_AEPD_new, on='Features')
final_weights
```
